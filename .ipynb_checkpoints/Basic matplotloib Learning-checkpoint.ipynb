```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,5,1)
y=x**3
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,1].plot(x,y)
ax[0,1].set(title="a simple plot",
           xlabel="X",
           ylabel="Y")
ax[1,0].scatter(x,y);
```

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(221)
ax1.plot(x,y)
ax2=fig.add_subplot(223)
ax2.scatter(x,y)
ax1.set(title="a simple plot",
           xlabel="X",
           ylabel="Y")
ax3=fig.add_subplot(224)
ax3.bar(x,y)
#plt.tight_layout()
```

```python
ax = fig.add_subplot(221)
ax.plot(x,y)
plt.show()
```
